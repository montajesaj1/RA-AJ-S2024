```python
import pandas as pd
```

```python
data = pd.read_csv('../margaret_scraper/task_2_pub_identified_60s.csv')
margaret = pd.read_csv('margaret_with_id.csv')
```

```python
# Assuming df_small has 15 columns initially, you would find the matching columns as follows

columns_to_match = [
    'url', 'common', 'preferred', 'brand', 'inn', 'pubchem', 'who_atc', 'originator', 
    'stereochemistry', 'formula', 'investigational', 'us_market', 'us_market_year', 'us_prev_market'
]

data['who_atc'] = data['who_atc'].astype(str)
margaret['who_atc'] = margaret['who_atc'].astype(str)

data['inn'] = data['inn'].astype(str)
margaret['inn'] = margaret['inn'].astype(str)

# Perform the merge to add unique_id to df_small
df_merged = pd.merge(data, margaret[['id'] + columns_to_match], on=columns_to_match, how='left')

df_merged
```

```python
df_merged.to_csv('task2_pub_60s_merged_id.csv', index=False)
```
