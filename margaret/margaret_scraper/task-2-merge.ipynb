```python
import pandas as pd 
margaret = pd.read_csv('unmatched_valid_margaret.csv')
margaret_num = margaret[margaret['pubchem'].notna()]
margaret_no_num = margaret[margaret['pubchem'].isna()]
```

```python
num = pd.read_csv('task_2_unamtched_valid_pub_num_cas.csv')
no_num = pd.read_csv('task_2_unamtched_valid_cas.csv')
```

```python
num
```

```python
no_num
```

```python
margaret_num.reset_index(drop=True, inplace=True)
num.reset_index(drop=True, inplace=True)

# Concatenating along columns
merged_num = pd.concat([margaret_num, num], axis=1)
merged_num
```

```python
margaret_no_num.reset_index(drop=True, inplace=True)
no_num.reset_index(drop=True, inplace=True)

# Concatenating along columns
merged_no_num = pd.concat([margaret_no_num, no_num], axis=1)
merged_no_num
```

```python
merged_no_num.to_csv('margaret-task2-no-pub.csv', index=False)
merged_num.to_csv('margaret-task2-with-pub.csv', index=False)
```
