```python
import selenium 
import csv
import re
import pandas as pd
import ast
import numpy as np
import math

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException
from selenium.webdriver.chrome.options import Options

import time
```

```python
margaret_identified = pd.read_csv('margaret_missing_cas_identified.csv')
filtered = margaret_identified[margaret_identified['pubchem'].notna()]
```

```python
filtered['pubchem'] = filtered['pubchem'].astype('int64')
pubchem_cid_arr = filtered['pubchem'].to_list()
```

```python
def get_pubchem_cid(cid):
    return f'https://pubchem.ncbi.nlm.nih.gov/compound/{cid}'
```

```python
def setup_webdriver():
    chrome_options = Options()
    chrome_options.add_argument("--headless")
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=chrome_options)
    return driver

def get_best_match(cid, driver):
    urls = []  # List to store URLs

    driver.get(get_pubchem_cid(cid))
    try:
        WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CSS_SELECTOR, '#Component-Compounds')))
        elements = driver.find_elements(By.CSS_SELECTOR, 'section#Component-Compounds a[data-ga-label="Content Link: Component-Compounds"]')
        for element in elements:
            url = element.get_attribute('href')
            urls.append(url)
    except (NoSuchElementException, TimeoutException):
        pass  # URLs and result_types remain empty if no match is found or if there's a timeout

    return urls

def get_best_matches(pubchem_cid_arr):
    urls = []
    driver = setup_webdriver()  # Initialize the WebDriver once
    try:
        for cid in pubchem_cid_arr:
            urls.append(get_best_match(cid, driver))
    finally:
        driver.quit()  # Make sure to quit the WebDriver

    return urls
```

```python
components = get_best_matches(pubchem_cid_arr)
```

```python
filtered['components'] = components
filtered.head(10)
```

```python
filtered.to_csv('margaret_with_components.csv', index=False)
```
