```python
import pandas as pd
```

```python
inxight = pd.read_csv('3_inxight_data_final.csv')
inxight.head()
```

```python
inxight_no_missing = inxight.dropna(subset=['phases'])
inxight_no_missing
```

```python
inxight_no_missing['conditions'] = inxight_no_missing['conditions'].str.split('; ')
inxight_no_missing['phases'] = inxight_no_missing['phases'].str.split('; ')

# Create a list of dictionaries to expand conditions and phases
expanded_data = []
for index, row in inxight_no_missing.iterrows():
    for condition, phase in zip(row['conditions'], row['phases']):
        expanded_row = row.copy()  # Make a copy of the original row
        expanded_row['conditions'] = condition.strip()
        expanded_row['phases'] = phase.strip()
        expanded_data.append(expanded_row)

# Create a new DataFrame from the expanded data
expanded_inxight_no_missing = pd.DataFrame(expanded_data)
expanded_inxight_no_missing
```

```python
final_inxight = expanded_inxight_no_missing[expanded_inxight_no_missing['phases'].isin(['Approved', 'Withdrawn'])]
final_inxight
```

```python
final_inxight.to_csv('inxight_to_match.csv', index=False)
```
