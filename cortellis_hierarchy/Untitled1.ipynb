```python
data = pd.read_csv("sample.csv")
```

```python
import ast
```

```python
def remove_brackets(column):
    new_column = []
    for item in column:
        # Convert string representation of list to actual list
        item_list = ast.literal_eval(item)
        
        # Remove outer two brackets if they exist
        if isinstance(item_list, list) and len(item_list) == 1:
            item_list = item_list[0]
            if isinstance(item_list, list) and len(item_list) == 1:
                item_list = item_list[0]
        
        # Convert back to string for storage in DataFrame
        new_column.append(str(item_list))
    return new_column

# Apply the function to the 'Cortellis Matches' and 'Aggregates' columns
data['Cortellis Matches'] = remove_brackets(data['Cortellis Matches'])
data['Aggregates'] = remove_brackets(data['Aggregates'])
```
